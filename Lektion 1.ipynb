Missing values
    - Hvad gør vi når vi skal se hvor der mangler "entries"
Checking for missing values in the dataset:
df.isnull().sum()
Den her måde får vi alle rækker hvor der er data:
df[~df.isnull()]

df.dropna() fjerner Na entries

Example: Filling missing values in the 'num_reviews' column with the average num_reviews.
df.fillna(df.number_reviews.mean(), inplace = True)

df.fillna(method = 'ffill', inplace = True) betyder at man bruger den forrige til at give dataet fra denne videre.
df.fillna(method = 'bfill', inplace = True) betyder at man bruger den efterfølgende værdi i den forrige til at udfylde Na
